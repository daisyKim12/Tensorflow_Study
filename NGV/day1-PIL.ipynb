## PIL
- Python Imaging Library
- Pillow라는 이름으로 포크(fork)되어 현재까지 지속 중
- 이미지 파일 열기, 닫기, 변환 등 기본적인 기능 제공

### Open Image

Opening the image from folder

```python
from PIL import Image
import requests

img = Image.open(requests.get('https://thumbs.dreamstime.com/z/red-ballon-party-birthday-colorful-color-30956951.jpg?w=576', stream=True).raw)
img
```

Search image format & size  
`size`method will return vertical, horizontal size in tuple

```python
print(img.format)
print(img.size)
```

Image mode  
Mode tells what kind of **color** mode image has
- RGB
- RGBA: A is for alpha it represents transparency
- L: Black and White and L represents Luminus
- CMYK: Represents color with cyan, megenta, yello, black and used for printing

```python
img.mode
```

### Image resize
Use `resize` method to resize images. The input must be in tuple

```python
img = img.resize((256,256))
```

### Image Augmentation
Vertical flop

```python
img. transpose(Image.FLIP_TOP_BOTTOM)
```

Rotate

```python
img.rotate(45)
```

Convert into Black and White

```python
img.convert('L')
```

### Split RGB 3 channel
Color images are consists of 3 channel.

```python
red, green, blue = img.split()
```

__Caution__ : Each channel only has L value so it will be Black and White when print red.

```python
red
```

### Using Empty Image to augemnt Image
Make a L image with same size and merge to change color.

```python
empty = Image.new('L', img.size)
Image.merge('RGB', (red, green, empty))
```

### Filter
Using filter in PIL we can extract outline of the image

```python
from PIL import ImageFilter
img1 = img.convert('L')
img2 = img1.filter(ImageFilter.FIND_EDGES)
img2
```

Emphsize outline in a image

```python
img3 = img2.filter(ImageFilter.EDGE_ENHANCE)
img3
```

Blurring a image

```python
img.filter(ImageFilter.BLUR)
```

Only the outline

```python
img.filter(ImageFilter.CONTOUR)
```
